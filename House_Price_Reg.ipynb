```python
import pandas as pd 
import numpy as np

df = pd.read_csv('kaggle_london_house_price_data.csv')
df

# Making a predicitve model using supervised learning
```

```python
df_1 = df.drop(columns=['outcode', 'saleEstimate_ingestedAt', 'fullAddress', 
                        'postcode', 'country',], axis=1)
df_1.columns
```

```python
df_1.head()
```

```python
df_2 = df_1.dropna()
df_2.isnull().sum()
```

```python
df_2.duplicated().sum()
```

```python
df_3 = df_2.drop_duplicates()
```

```python
df_3.duplicated().sum()
```

```python
from sklearn import preprocessing

label_encode = preprocessing.LabelEncoder()

df_3 = df_3.copy()

df_3['saleEstimate_confidenceLevel'] = label_encode.fit_transform(df_3['saleEstimate_confidenceLevel'])
```

```python
len(df_3)
```

```python
train_data_1, test_data, val_data_1 = df_3[:140001], df_3[140001:150001], df_3[150001:]

train_data = train_data_1[['saleEstimate_confidenceLevel',  
       'rentEstimate_lowerPrice', 'rentEstimate_currentPrice',
       'rentEstimate_upperPrice', 'saleEstimate_lowerPrice',
       'saleEstimate_currentPrice', 'saleEstimate_upperPrice',
       'saleEstimate_valueChange.numericChange',
       'saleEstimate_valueChange.percentageChange', 'history_price',
       'history_percentageChange', 'history_numericChange']]

val_data = val_data_1[['saleEstimate_confidenceLevel',  
       'rentEstimate_lowerPrice', 'rentEstimate_currentPrice',
       'rentEstimate_upperPrice', 'saleEstimate_lowerPrice',
       'saleEstimate_currentPrice', 'saleEstimate_upperPrice',
       'saleEstimate_valueChange.numericChange',
       'saleEstimate_valueChange.percentageChange', 'history_price',
       'history_percentageChange', 'history_numericChange']]


# Remember to use '[[]]'
```

```python
train_data
```

```python
X_train, y_train = train_data.to_numpy()[:, :-1], train_data.to_numpy()[:, -1]
X_val, y_val = val_data.to_numpy()[:, :-1], val_data.to_numpy()[:, -1]
X_train
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train[:, 3:])

def pre_pro(X):
    a = np.copy(X)
    a[:, 3:] = scaler.transform(a[:, 3:])
    return a


X_train_process = pre_pro(X_train)
X_train_process

X_val_process = pre_pro(X_val)
X_val_process
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(X_train_process, y_train)
rse = mse(lr.predict(X_train_process), y_train, squared=False)
rse_val = mse(lr.predict(X_val_process), y_val, squared=False)
rse, rse_val
# RMSE measures how well the model fits the training data
# A lower RMSE value means the model’s predictions are closer to the actual values, indicating better accuracy
# rse_val measures how well the model generalises to unseen validation data
# If rse_val is much larger than rse, it might suggest the model is overfitting the training data.
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=3).fit(X_train_process, y_train)

rse = mse(rf.predict(X_train_process), y_train, squared=False)
rse_val = mse(rf.predict(X_val_process), y_val, squared=False)
rse, rse_val
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=2).fit(X_train_process, y_train)

rse = mse(knn.predict(X_train_process), y_train, squared=False)
rse_val = mse(knn.predict(X_val_process), y_val, squared=False)
rse, rse_val

# KNN 'n_neighbors=' cannot be 1
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(n_estimators=50).fit(X_train_process, y_train)

rse = mse(gb.predict(X_train_process), y_train, squared=False)
rse_val = mse(gb.predict(X_val_process), y_val, squared=False)
rse, rse_val
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

simple_ann = Sequential()
# Defining the layers
simple_ann.add(Input((11,))) # dataset has 11 features to input
simple_ann.add(Dense(2, 'relu'))
simple_ann.add(Dense(1, 'linear'))

opt = Adam(learning_rate=0.1)
cp = ModelCheckpoint('models/simple_ann.keras', save_best_only=True)
simple_ann.compile(optimizer=opt, loss='mse', metrics=[RootMeanSquaredError()])
simple_ann.fit(x=X_train_process, y=y_train, validation_data=(X_val_process, y_val), callbacks=[cp], epochs=10)

# loss: the Mean Squared Error (MSE) (loss='mse' setting)
# Measures the average squared difference between your model’s predicted values and the true target values. Lower values indicate better fit.

# root_mean_squared_error: measures the average prediction error in the original units of the target variable
# Lower values mean the predictions are closer to the actual values on average

# val_loss: This is the loss (MSE) calculated on the validation data
# Evaluates how well the model generalises to new data

# val_root_mean_squared_error: This is the RMSE calculated on the validation data
# Also, shows the typical prediction error on unseen data.
```

```python
from tensorflow.keras.models import load_model

simple_ann = load_model('models/simple_ann.keras')
mse(simple_ann.predict(X_train_process), y_train, squared=False), mse(simple_ann.predict(X_val_process), y_val, squared=False)
```

```python
compl_ann = Sequential()
compl_ann.add(Input((11,))) # dataset has 11 features to input
compl_ann.add(Dense(32, 'relu'))
compl_ann.add(Dense(16, 'relu'))
compl_ann.add(Dense(1, 'linear'))

opt = Adam(learning_rate=0.1)
cp = ModelCheckpoint('models/compl_ann.keras', save_best_only=True)
compl_ann.compile(optimizer=opt, loss='mse', metrics=[RootMeanSquaredError()])
compl_ann.fit(x=X_train_process, y=y_train, validation_data=(X_val_process, y_val), callbacks=[cp], epochs=10)
```

```python
from tensorflow.keras.models import load_model

compl_ann = load_model('models/compl_ann.keras')
mse(compl_ann.predict(X_train_process), y_train, squared=False), mse(simple_ann.predict(X_val_process), y_val, squared=False)
```
